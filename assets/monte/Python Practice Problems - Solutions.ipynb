1. Write a funciton that calculates the present value of a single, future cash flow: $$ PV_0 = \frac{FV_t}{(1+r)^t} $$

```python
def pv(r,t,fv):
    pv = fv/(1+r)**t
    return pv
```

```python
print(pv(.10,1,110))
```

2. Write a funciton that calculates the present value of an annuity: $$ PV_0 = \sum_{t=1}^{N} \frac{PMT_t}{(1+r)^t} $$

```python
def pv_annuity(r,n,pmt):
    pva = 0  
    for t in range(n):
        pva = pva + pv(r,t+1,pmt)
    return pva
```

```python
print(pv_annuity(.10,2,100))
```

4. Write a program to print the Fibonacci series between 0 to 50. The Fibonacci Sequence is the series of numbers 0, 1, 1, 2, 3, 5, 8, 13, 21, etc. Every next number is found by adding up the two numbers before it.

```python
x = 0
y = 1
for i in range(10):
    y_new = x + y
    x_new = y
    if y_new < 50:
        print(y_new)
    x = x_new
    y = y_new
```

```python
def convert_temp(f):
    c = 1.8*f+32
    return c
```

7. Write a function to sum all the numbers in a list.

```python
def sum_list(x):
    y = 0
    for i in x:
        y = y + i
    return y
```

```python
x = [5,2,3]
print(sum_list(x))
```
